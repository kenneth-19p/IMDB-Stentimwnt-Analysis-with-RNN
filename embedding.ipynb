```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
## Define the vocabulary size

voc_size = 10000
```

```python
### One Hot Representation

one_hot_repre = [one_hot(words, voc_size) for words in sent]
one_hot_repre
```

```python
## Word Embedding Representation

from tensorflow.keras.layers import Embedding, InputLayer
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
## Since the sentences are not of equal lenght, we set the max sentence lenght to add zeros to all the sentences for RNN
sent_lenght = 8
embedded_docks = pad_sequences(one_hot_repre,padding='pre', maxlen=sent_lenght)
embedded_docks
```

```python
### Feature representation
dim=10
```

```python

model = Sequential()
model.add(InputLayer(input_shape=(sent_lenght,)))
model.add(Embedding(voc_size, dim, input_length=sent_lenght))
model.compile(optimizer='adam', loss='mse')


```

```python
model.summary()
```

```python
model.predict(embedded_docks)
```
